# LabelEncoder

```python
import pandas as pd

train = pd.read_csv('./train.csv')
train.head()
```

# 1. astype('category').cat.codes

    - numerical 데이터로 변경하고 싶은 categorical컬럼을 정함
    - categorical 형 column으로 강제 형변환
    - cat.codes를 호출하면 자동으로 숫자형 리턴

```python
train['Sex'].astype('category').cat.codes
```

# 2. LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
result = le.fit_transform(train['Sex'])
print(result)
```

```python
le.inverse_transform(result)
```

# 3. get_dummies

```python
train['Sex'].astype('category').cat.codes
```

```python
pd.get_dummies(train['Sex'])
```

### Numerical -> Categorical로 변환

```python
train['Pclass'].astype('category').cat.codes
```

```python

pd.get_dummies(train['Pclass'])
```
